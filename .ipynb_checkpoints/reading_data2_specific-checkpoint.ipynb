# Objective
### The aim of this notebook is to read the volunteer scan data imported from the lab server, for subject 2
First read the data from the scan_data folder and then combine to generate a big data file. Then move forward as in the t2shpics notebook.

The outputs are saved so can be easily used in the NN fit notebook. 

Generate mask for fully sampled and under sampled reconstruction, 

# Keep separate folders for data of different subjects so that there is no funky buisness happening while processing data of different users


```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from bart import bart#required for importing bart

import cfl #useful in data storage and reading, dont need it right now
```

```python
import os
print(os.getcwd())#get current working directory
os.chdir('/home/ubuntu/sid_notebooks/')#change current working directory
print(os.getcwd())
```

```python
data860 = np.asarray(cfl.readcfl('scan_data/subject2_data860'))
data1830 = np.asarray(cfl.readcfl('scan_data/subject2_data1830'))
data2800 = np.asarray(cfl.readcfl('scan_data/subject2_data2800'))
print(data860.shape)
print(data1830.shape)
print(data2800.shape)
```

```python
#data860 have only one slice in the data, so just append 0 there for right now, we will not use it later anyways
data860 = np.concatenate((np.zeros((data860.shape)),data860),axis=4)
print(data860.shape)
data_full = np.concatenate((data860,data1830,data2800),axis=7)
print(data_full.shape)
```

```python
data_full = data_full.squeeze()
print(data_full.shape)
data_full = np.moveaxis(data_full, [2, 1], [1, 2])#swap axises to bring X and  Y in first 2 dimensions
print(data_full.shape)#0,1 for x and y dimension, 2 for coils, 3 for slice index, 4 for the ETL
data_full = np.rot90(data_full)#rotate the image 90 to bring it into the correct orientation
print(data_full.shape)
```

```python
# only taking the even echoes
data_full = data_full[:,:,:,:,1::2]
print(data_full.shape)
```

```python
data_fft = bart(1,'fft -u -i 3',data_full)#going to the k-space for all the coil images
print(data_fft.shape)
```

```python
#generating masks for fully sampling the k-space domain
mask_temp = bart(1,'t2sh_gen_mask -Y 288 -Z 288 -d 1 -C 3 -V 5 -r 1 -T 1 -E 1 -F')
#will only sample the circular patch in middle 
print(mask_temp.shape)
plt.imshow(np.abs(mask_temp[0,:,:]).squeeze(), cmap='gray',vmax = 0.1,vmin=0)
plt.title('Figure: Fully sampled k-space')
plt.show()
```

```python
# basis from the experimental T1 and T2 values obtained using dictionary learning
X_sim = np.load('subject2/X_simulated_experimentalT1_T2.npy')#load the simulated epg curves for 390 TEs and mask2 of the data
print(X_sim.shape)
basis = np.load('subject2/basis_experimentalT1_T2.npy')#load the basis matrix
print(basis.shape)
bas = np.load('subject2/basis_experimentalT1_T2.npy')#loading the basis
print(bas.shape)
bas = bas[None,None,None,None,None,:,:]#bringing it to the required dimensions
print(bas.shape)
#cfl.writecfl('bas',bas)
```

```python
brain_phantom = data_fft[None,:,:,:,1,None,:]#choosing the 2nd slice becase for subject 2, first slice is not done for 860
print(brain_phantom.shape)
brain_phantom_fft = bart(1,'fft -u 6',brain_phantom)#going to the k-space for all the coil images
```

```python
for index in range(8):
    plt.figure(1, figsize=(20,5))
    plt.subplot(1, 8, index+1)
    plt.imshow(np.abs(brain_phantom[0,:,:,index,0,1]),cmap='gray', vmin=0, vmax=.00005)
    plt.title('coil {}'.format(index+1))

plt.show()
print('The brain images as acquired by 8 different coils')
```

```python
sens = bart(1, 'ecalib -t .0002 -c 0.5 -m1', brain_phantom_fft).squeeze()#some parameters need to be optimized for the ecalib algorithm but
sens = sens[None,:,:,:] #bring to the required format
```

```python
plt.figure(figsize=(16,20))
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(abs(sens[0,:,:,i]).squeeze(), cmap='gray')
    plt.title('Coil sensitivity {}'.format(i))
plt.show()
print('The coil sensitivities as estimated by the ecalib algorithm')
```

```python
#generating k-space data for the fully sampled case with 8 coils
print(brain_phantom_fft.shape)
print(mask_temp.shape)
#fully_sampled_fft = brain_phantom_fft
fully_sampled_fft = bart(1,'fmac',  brain_phantom_fft,mask_temp)#sample using the fully sampled mask
print(fully_sampled_fft.shape)
print(sens.shape)
```

```python
l2 = 0.001#regularization parameter
pics_out = bart(1,'pics -R Q:0.005 -i 10 -S',brain_phantom_fft,sens)

pics_out1=pics_out.squeeze()#squeeze out the necessary dimensions
```

```python
pics_out_coeffs = bart(1,'t2sh_proj -K 5',pics_out,bas)
print(pics_out_coeffs.shape)

pics_out1=pics_out.squeeze()#squeeze out the necessary dimensions
```

```python
pics_out1=pics_out.squeeze()#squeeze out the necessary dimensions
plt.figure(figsize=(16,12))
for index in range(4):
    plt.subplot(2,4,index+1)
    plt.imshow(np.abs(pics_out1[:,:,index*12]), cmap='gray',vmax = .0003,vmin=0)
    plt.title('Echo number: ' + str(12*index))
```

```python
#project the fully sampled data onto the basis with 5 basis vectors choosen
cfksp2 = bart(1,'t2sh_proj -K 5',fully_sampled_fft,bas)
cfksp2 = cfksp2.squeeze()
print(cfksp2.shape)
cfksp2 = cfksp2[None,:,:,:,None,None,:]
print('cfksp2: ',cfksp2.shape)
print('mask_temp: ',mask_temp.shape)
print('sensitivity: ',sens.shape)
print('basis: ',bas.shape)
```

```python
#reconstructing the fully sampled image through the t2sh recon command
#fully_image1 = bart(1,'t2sh_pics -R L:7:7:0.00005 -i 10 -K 5',cfksp2,mask_temp,sens,bas)#only 10 iterations here
#because it is coverging early
fully_image1 = bart(1,'t2sh_pics -i 10 -K 5',cfksp2,mask_temp,sens,bas)#only 10 iterations here

print(fully_image1.shape)

fully_image_se = bart(1,'t2sh_proj -K 5 -f',fully_image1,bas)#project back to get the image for all the TE values
```

```python
fully_image1=fully_image1.squeeze()#squeeze out the necessary dimensions
plt.figure(figsize=(16,12))
for index in range(5):
    plt.subplot(1,5,index+1)
    plt.imshow(np.abs(fully_image1[:,:,index]), cmap='gray',vmax = 0.0008,vmin=0)
    plt.title('Fully sampled Coeffs = ' + str(index))
    
    
pics_out_coeffs1 = pics_out_coeffs.squeeze()
plt.figure(figsize=(16,12))
for index in range(5):
    plt.subplot(1,5,index+1)
    plt.imshow(np.abs(pics_out_coeffs1[:,:,index]), cmap='gray',vmax = 0.0008,vmin=0)
    plt.title('Pics Coeffs = ' + str(index))
    
plt.figure(figsize=(16,12))
for index in range(5):
    plt.subplot(1,5,index+1)
    plt.imshow(np.abs(fully_image1[:,:,index] - pics_out_coeffs1[:,:,index]), cmap='gray',vmax = 0.0008*0.01,vmin=0)
    plt.title('Diff Coeffs = ' + str(index))
```

```python
fully_image_se1=fully_image_se.squeeze()#squeeze out the necessary dimensions
```

```python
error_pics = np.linalg.norm(pics_out1 - fully_image_se1) / np.linalg.norm(pics_out1)
print('nrmse between pics images and T2sh_pics images:- ',error_pics)
```

## Now starting to reconstruct the undersampled data

```python
exam_time = 7*60#in seconds, total exam time
TR_ratio = np.array([0.34,0.4,0.26])#ratio of different TR values for exam
TR_vals = np.array([860,1830,2800])#TR values given 
number_of_TRs = np.floor(exam_time*TR_ratio*1000/TR_vals)
print(number_of_TRs)
```

```python
mask1 = bart(1,'t2sh_gen_mask -Y 288 -Z 288 -d 130 -C 3 -V 5 -r 1 -T 166 -E 16')
#the mask generates a total 256X256 k-space where the number of sampling masks is 130 the variational density is 5, 
# the central calibration region is of 3, random seed is 1, number of echo trains is 158 and echo train length is 130

mask2 = bart(1,'t2sh_gen_mask -Y 288 -Z 288 -d 130 -C 3 -V 5 -r 2 -T 91 -E 16')

mask3 = bart(1,'t2sh_gen_mask -Y 288 -Z 288 -d 130 -C 3 -V 5 -r 3 -T 39 -E 16')

mask_full = np.concatenate((mask1,mask2,mask3),axis=5)
print(mask_full.shape)
```

```python
plt.imshow(np.abs(mask_full[0,:,:,0,0,30]).squeeze(), cmap='gray',vmax = 0.1,vmin=0)
plt.title('Figure: undersampled sampled k-space')
plt.show()
```

```python
print(fully_sampled_fft.shape)
print(mask_full.shape)
```

```python
#generating the k-space for the undersampled data
teksp = bart(1,'fmac',fully_sampled_fft,mask_full)#generating the k-space data for all the different coils
teksp.shape
```

```python
cfksp = bart(1,'t2sh_proj -K 5',teksp,bas)#project the teksp onto the basis with 5 basis vectors choosen
cfksp = cfksp.squeeze()
print(cfksp.shape)
```

```python
cfksp1 = cfksp[None,:,:,:,None,None,:]
print('cfksp: ',cfksp1.shape)
print('basis: ',bas.shape)
print('mask: ',mask_full.shape)
print('sensiticity shape: ',sens.shape)
maps_final = sens
```

```python
#recon2 = bart(1,'t2sh_pics -i 250 -K 5',cfksp1,mask_full,maps_final,bas)#without LLR
recon2 = bart(1,'t2sh_pics -R L:7:7:0.00005 -i 250 -K 5',cfksp1,mask_full,maps_final,bas)
```

```python
recon_image_sc = bart(1,'t2sh_proj -K 5 -f',recon2,bas)#project back to get the image for all the TE values
```

```python
plt.figure(figsize=(16,12))
for index in range(4):
    plt.subplot(1,8,index+1)
    plt.imshow(np.abs(fully_image_se1[:,:,index*8]), cmap='gray',vmax = 0.00005,vmin=0)
    plt.title('Echo index: ' + str(8*index))
plt.show()
print('Fully sampled Images generated using t2sh command')

recon_image_sc1=recon_image_sc.squeeze()#squeeze out the necessary dimensions
plt.figure(figsize=(16,12))
for index in range(4):
    plt.subplot(1,8,index+1)
    plt.imshow(np.abs(recon_image_sc1[:,:,index*8]), cmap='gray',vmax = 0.00005,vmin=0)
    plt.title('Echo index: ' + str(8*index))
plt.show()
print('Images generated from the t2sh recon command') 
```

```python
recon1=recon2.squeeze()
plt.figure(figsize=(16,12))
for index in range(3):
    plt.subplot(1,5,index+1)
    plt.imshow(np.abs(recon1[:,:,index]), cmap='gray')#,vmax = .00008,vmin=0)
    plt.title('Reconstructed Coeffs = ' + str(index))
    
fully_image1 = fully_image1.squeeze()
plt.figure(figsize=(16,12))
for index in range(3):
    plt.subplot(1,5,index+1)
    plt.imshow(np.abs(fully_image1[:,:,index]), cmap='gray')#,vmax = 800000,vmin=0)
    plt.title('FS Coeffs = ' + str(index))
```

```python
c_linear_coeffs_T1 = np.load('subject2/96ETL_c_linear_coeffs_T1.npy')
c_linear_coeffs_T2 = np.load('subject2/96ETL_c_linear_coeffs_T2.npy')
c_linear_coeffs_T1FLAIR = np.load('subject2/96ETL_c_linear_coeffs_T1FLAIR.npy')#load the saved linear coefficients
c_linear_coeffs_T2FLAIR = np.load('subject2/96ETL_c_linear_coeffs_T2FLAIR.npy')#
print(c_linear_coeffs_T2FLAIR)
```

```python
K = 5#No. of basis coefficients to be used for linear combination
```

```python
fully_image2 = fully_image1[:,:,:K]
T1_image_FS_LC = np.dot(fully_image2,c_linear_coeffs_T1[:K])
T2_image_FS_LC = np.dot(fully_image2,c_linear_coeffs_T2[:K])
T1_flair_FS_LC = np.dot(fully_image2,c_linear_coeffs_T1FLAIR[:K])
T2_flair_FS_LC = np.dot(fully_image2,c_linear_coeffs_T2FLAIR[:K])

plt.figure(figsize=(16,12))
plt.subplot(1,4,1)
plt.imshow(np.abs(T1_image_FS_LC[:,:]), cmap='gray',vmax = 0.0003,vmin=0)
plt.title('T1 weighted image')

plt.subplot(1,4,2)
plt.imshow(np.abs(T2_image_FS_LC[:,:]), cmap='gray',vmax = 0.0005,vmin=0)
plt.title('T2 weighted image')

plt.subplot(1,4,3)
plt.imshow(np.abs(T1_flair_FS_LC[:,:]), cmap='gray',vmax = 0.0005,vmin=0)
plt.title('T1 FLAIR image')

plt.subplot(1,4,4)
plt.imshow(np.abs(T2_flair_FS_LC[:,:]), cmap='gray',vmax = 0.0003,vmin=0)
plt.title('T2 FLAIR image')

plt.show()
print('     Fully sampled linear combination (K = {})'.format(K))
```

```python
pics_out_coeffs2 = pics_out_coeffs1[:,:,:K]
T1_image_FS_LC = np.dot(pics_out_coeffs2,c_linear_coeffs_T1[:K])
T2_image_FS_LC = np.dot(pics_out_coeffs2,c_linear_coeffs_T2[:K])
T1_flair_FS_LC = np.dot(pics_out_coeffs2,c_linear_coeffs_T1FLAIR[:K])
T2_flair_FS_LC = np.dot(pics_out_coeffs2,c_linear_coeffs_T2FLAIR[:K])

plt.figure(figsize=(16,12))
plt.subplot(1,4,1)
plt.imshow(np.abs(T1_image_FS_LC[:,:]), cmap='gray',vmax = 0.0003,vmin=0)
plt.title('T1 weighted image')

plt.subplot(1,4,2)
plt.imshow(np.abs(T2_image_FS_LC[:,:]), cmap='gray',vmax = 0.0005,vmin=0)
plt.title('T2 weighted image')

plt.subplot(1,4,3)
plt.imshow(np.abs(T1_flair_FS_LC[:,:]), cmap='gray',vmax = 0.0005,vmin=0)
plt.title('T1 FLAIR image')

plt.subplot(1,4,4)
plt.imshow(np.abs(T2_flair_FS_LC[:,:]), cmap='gray',vmax = 0.0003,vmin=0)
plt.title('T2 FLAIR image')

plt.show()
print('     Pics coeffs linear combination (K = {})'.format(K))
```

```python
recon3 = recon1[:,:,:K]
T1_image_reconstructed = np.dot(recon3,c_linear_coeffs_T1[:K])
T2_image_reconstructed = np.dot(recon3,c_linear_coeffs_T2[:K])
T1_flair_reconstructed = np.dot(recon3,c_linear_coeffs_T1FLAIR[:K])
T2_flair_reconstructed = np.dot(recon3,c_linear_coeffs_T2FLAIR[:K])

plt.figure(figsize=(16,12))
plt.subplot(1,4,1)
plt.imshow(np.abs(T1_image_FS_LC[:,:]), cmap='gray',vmax = .001,vmin=0)
plt.title('T1 weighted image')

plt.subplot(1,4,2)
plt.imshow(np.abs(T2_image_FS_LC[:,:]), cmap='gray',vmax = 0.001,vmin=0)
plt.title('T2 weighted image')

plt.subplot(1,4,3)
plt.imshow(np.abs(T1_flair_FS_LC[:,:]), cmap='gray',vmax = 0.001,vmin=0)
plt.title('T1 FLAIR image')

plt.subplot(1,4,4)
plt.imshow(np.abs(T2_flair_FS_LC[:,:]), cmap='gray',vmax = 0.0005,vmin=0)
plt.title('T2 FLAIR image')

plt.show()
print('     Reconstructed images with linear combination (K = {})'.format(K))
```

```python
def nrmse(x1, x2):
    return np.linalg.norm(x1.ravel() - x2.ravel()) / np.linalg.norm(x1.ravel())
```

```python
print(nrmse(np.abs(T1_image_FS_LC),np.abs(True_contrast_T1_image)))
print(nrmse(np.abs(T2_image_FS_LC),True_contrast_T1_image))
print(nrmse(np.abs(T1_flair_FS_LC),True_contrast_T1_FLAIR))
print(nrmse(np.abs(T2_flair_FS_LC),True_contrast_T2_FLAIR))
```
